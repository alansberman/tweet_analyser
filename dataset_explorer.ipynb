```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\alan\\Desktop\\trainingandtestdata\\training.csv",encoding="latin-1")
# Get only valence and tweet
df= df[['Valence','Tweet']]
# Replace target (Valence) values so that they're 0 and 1
df["Valence"]=df["Valence"].replace([4],1)
```

```python

# Remove URLs and @ tags
# Thanks to  https://stackoverflow.com/questions/11331982/how-to-remove-any-url-within-a-string-in-python
df['Tweet']=df['Tweet'].str.replace(pat='https?:\/\/.*[\r\n]*',repl='',regex=True)
df['Tweet']=df['Tweet'].str.replace(pat="\S+.com\s",repl="",regex=True)
df['Tweet']=df['Tweet'].str.replace(pat="@[A-Za-z0-9_]+\s",repl="",regex=True)

df.to_csv("C:\\Users\\alan\\Desktop\\trainingandtestdata\\training_data.csv")
```
